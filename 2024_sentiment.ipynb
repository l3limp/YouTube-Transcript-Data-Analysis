```python
import pandas as pd
from pprint import pprint
import os
import matplotlib.pyplot as plt

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('mkbhd.csv')
```

```python
df = df[df['Year'] == 2024]
all_2024_scripts = df['Captions'].tolist()
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
def getSentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(text)
    return scores
```

```python
filteredScripts = []

for script in all_2024_scripts:

      d = nlp(script)
      script_annotated = [tok.text for tok in d if not
                                  (tok.is_punct or tok.is_stop 
                                    or tok.is_space or tok.pos_ == "PROPN" )
                                  ]                         
      script_annotated_string = ' '.join(script_annotated)                         
      filteredScripts.append(script_annotated_string)
```

```python
df['Filtered Scripts'] = filteredScripts
df.head(14)
```

```python
scatterDf = pd.DataFrame(columns=['Id', 'Positive', 'Negative', 'Neutral'])


scores = {}
for row in df.values:
    #row[9] is filtered scripts
    scores[row[1]] = getSentiment(row[12])

for key, value in scores.items():
    scatterDf.loc[len(scatterDf.index)] = [key, value['pos'], value['neg'], value['neu']]
```
